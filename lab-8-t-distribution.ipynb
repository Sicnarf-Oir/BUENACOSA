```python
# imports and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# load dataset
df = pd.read_csv(r'transformer-voltage-10-sample.csv', 
                 delimiter=",",
                 index_col='Sample')
df
```

```python
# summary of dataframe
df.info()
```

```python
# summary of statistics
df.describe()
```

```python
# t-critical for two-tailed tes
alpha = 0.05
n = df['Voltage'].count()

t_critical = stats.t.ppf(1-alpha/2,n-1)
t_critical
```

```python
# confidence interval
n = df['Voltage'].count()
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

ci_lower, ci_upper = stats.t.interval(
    confidence = 0.95,
    df = n -1,
    loc = mean,
    scale = std/np.sqrt(n)
)

print("CI =", "[",  ci_lower, ",", ci_upper, "]")
```

```python
# voltage distribution
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

# x-axis
x_min = df['Voltage'].min()
x_max = df['Voltage'].max()

# normal distribution plot
x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
        color="#E21414",
        linewidth=2.9)

# histogram
plt.hist(df['Voltage'], bins=5,
        density=True,
        alpha=0.8,
        color="#E27B14",
        edgecolor="white")

plt.title('Normal Distribution of Current Values',
            font = 'Cambria',
            fontsize = 12)
plt.ylabel('Probability Density')
plt.xlabel('Voltage (V)')

# confidence interval
plt.axvline(x = ci_lower,
            color='Black',
            linestyle='--')
plt.axvline(x = ci_upper,
            color='Black',
            linestyle='--')
plt.show()
```

# Shoe Inventory

```python
# load dataset
df = pd.read_csv(r'shoe-inventory.csv', 
                 delimiter=",",
                 )
df
```

```python
# (rows and columns)
df.shape
```

```python
# mean of size 6
mean_size_6 = df.iloc[0,1:13].mean()
mean_size_6
```

```python
# standard deviation of size 6
std_size_6 = df.iloc[0,1:13].std()
std_size_6
```

```python
# size 6 95% CI
n = df.shape[1]-1 # returns 13 in (17, 13)

lower_6, upper_6 = stats.t.interval(
    confidence = 0.95,
    df = n-1,
    loc = mean_size_6,
    scale = std_size_6/np.sqrt(n)
)

print(f"CI_6:[{lower_6:.2F}, {upper_6:.2F}]")
```

```python
# 95% Confidence Level
mean = df.iloc[0:17, 1:13].mean(axis=1)
std = df.iloc[0:17, 1:13].std(axis=1)

df['ci_lower'], df['ci_upper'] = stats.t.interval(
    confidence = 0.95,
    df = n-1,
    loc = mean,
    scale = std/np.sqrt(n)
)
print(f"CI_6:[{ci_lower:.2F}, {ci_upper:.2F}]")
```
