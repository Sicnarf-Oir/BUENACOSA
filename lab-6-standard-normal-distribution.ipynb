```python
# imports and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

# Z - score

```python
# creating a dataframe
df = pd.DataFrame([1,2,2,3,3,3,4,4,5],
                    columns = ['Dataset'])
df
```

```python
# summary of statistics
df.describe()
```

```python
# Z-scores
df['Z-score'] = stats.zscore(df['Dataset'],ddof=1)
df
```

```python
df['Z-score'].mean()
```

```python
df['Z-score'].std(ddof=1)
```

```python
# load dataset
df = pd.read_csv(r'current-test.csv', 
                 delimiter=",",
                 index_col='Sample')
df
```

```python
df['Current'][2]
```

# Histogram

```python
# histogram plot
plt.hist(df['Current'], bins=5,
        alpha=0.8,
        color="#E27B14",
        edgecolor="red")

plt.title('Histogram of Current Values',
            font = 'Cambria',
            fontsize = 12)

plt.xlabel('Current (mA)')
plt.ylabel('Frequency')
plt.show()
```

# Normal Distribution

```python
# normal distribution curve
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

# x-axis
x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
        color="#E27B14",
        linewidth=2.9)

plt.title('Normal Distribution of Current Values',
            font = 'Cambria',
            fontsize = 12)

plt.xlabel('Current (mA)')
plt.ylabel('Probability Densityy')
plt.show()
```

```python
# histogram plot
plt.hist(df['Current'], bins=5,
        density=True,
        alpha=0.8,
        color="#E27B14",
        edgecolor="red")

plt.title('Normal Distribution of Current Values',
            font = 'Cambria',
            fontsize = 12)

# normal distribution curve
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
        color="blue",
        linewidth=2.9)
plt.show()
```

# Standard Normal Distribution

```python
# standard normal distribution

# x-axis
x_min = df['Z-score'].min()
x_max = df['Z-score'].max()
x = np.linspace(x_min, x_max, 100)

p = stats.norm.pdf(x, df['Z-score'].mean(), df['Z-score'].std(ddof=1))

plt.plot(x, p,
        color="#E27B14",
        linewidth=2.9)

plt.title('Standard Normal Distribution of Current Values N(0, 1)',
            font = 'Cambria',
            fontsize = 12)

plt.xlabel('Z-score')
plt.ylabel('Probability Densityy')
plt.show()
```
